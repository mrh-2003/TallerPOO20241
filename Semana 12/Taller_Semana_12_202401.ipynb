```python
class CuentaCorriente():
  __DNI = ""
  __nombre = ""
  __apellidos = ""
  __distrito =""
  __telefono = ""
  __saldo = 0.0
  def __init__(self, dni, nombre, apellidos, distrito, telefono, saldo):
    self.__DNI = dni
    self.__nombre = nombre
    self.__apellidos = apellidos
    self.__distrito = distrito
    self.__telefono = telefono
    self.__saldo = saldo
  def get_DNI(self):
    return self.__DNI
  def get_nombre(self):
    return self.__nombre
  def get_apellidos(self):
    return self.__apellidos
  def get_distrito(self):
    return self.__distrito
  def get_telefono(self):
    return self.__telefono
  def get_saldo(self):
    return self.__saldo
  def set_DNI(self, valor):
    self.__DNI = valor
  def set_nombre(self, valor):
    self.__nombre = valor
  def set_apellidos(self, valor):
    self.__apellidos = valor
  def set_distrito(self, valor):
    self.__distrito = valor
  def set_telefono(self, valor):
    self.__telefono = valor
  def set_saldo(self, valor):
    self.__saldo = valor
  def retirar_dinero(self, monto):
    self.__saldo -= monto
  def ingresar_dinero(self, monto):
    self.__saldo += monto
  def consultar_cuenta(self):
    print(f"DNI: {self.__DNI}\nNombre: {self.__nombre}\nApellidos: {self.__apellidos}\nDistrito: {self.__distrito}\nTelefono: {self.__telefono}\nSaldo: S/{self.__saldo}")
  def saldo_negativo(self):
    return self.__saldo < 0

class Controladora():
  __lista = []
  def agregar(self, cuenta):
    self.__lista.append(cuenta)
  def existe_cuenta(self, dni):
    for i in range(len(self.__lista)):
      if self.__lista[i].get_DNI() == dni:
        return i
    return -1

  def realizar_transferencia(self, dniOrigen, dniDestino, monto):
    indiceOrigen = self.existe_cuenta(dniOrigen)
    indiceDestino = self.existe_cuenta(dniDestino)
    if indiceOrigen == -1 or indiceDestino == -1 or indiceDestino == indiceOrigen:
      print("Error en realizar la transferencia")
    else:
      self.__lista[indiceOrigen].retirar_dinero(monto)
      self.__lista[indiceDestino].ingresar_dinero(monto)
  def saldo_promedio_por_distrito(self, distrito):
    suma = cont =  prom = 0
    for cuenta in self.__lista:
      if cuenta.get_distrito() == distrito:
        suma+= cuenta.get_saldo()
        cont+=1
    if cont > 0:
      prom = suma / cont
    print(f"El promedio del distrito {distrito} es {prom}")
  def buscar_personar_con_inicial(self, caracter):
    print("Personas con la inical", caracter)
    for cuenta in self.__lista:
      if cuenta.get_nombre()[0] == caracter:
        cuenta.consultar_cuenta()
  def personas_mayor_saldo(self):
    saldos = []
    for cuenta in self.__lista:
      saldos.append(cuenta.get_saldo())
    maxSaldo = max(saldos)
    print("Personas con mayor saldo: ")
    for cuenta in self.__lista:
      if cuenta.get_saldo() == maxSaldo:
        cuenta.consultar_cuenta()
  def consultar_cuentas(self):
    print("Listado de cuentas: ")
    for cuenta in self.__lista:
      cuenta.consultar_cuenta()
      print("Saldo negativo:", cuenta.saldo_negativo())
# main
contro = Controladora()
dni = " "
while dni != "":
  dni = " "
  while not(len(dni)==8 or dni == ""):
    dni = input("Ingrese DNI: ")
  if dni != "":
    nombre = ""
    while not(nombre != ""):
      nombre = input("Ingresar nombre: ")
    apellidos = input("Ingresar apellido: ")
    distrito = input("Ingresar distrito: ")
    telefono = input("Ingresar telefono: ")
    saldo = -1
    while not(saldo >= 0):
      saldo = float(input("Ingrese saldo: "))
    cuenta = CuentaCorriente(dni, nombre, apellidos, distrito, telefono, saldo)
    contro.agregar(cuenta)
contro.consultar_cuentas()
dniOrigen = ""
while not(len(dniOrigen) == 8):
  dniOrigen = input("Ingresar DNI Origen: ")
dniDestino = ""
while not(len(dniDestino) == 8):
  dniDestino = input("Ingresar DNI Descuento: ")
contro.realizar_transferencia(dniOrigen, dniDestino, 200)
contro.consultar_cuentas()
contro.saldo_promedio_por_distrito("Victoria")
contro.buscar_personar_con_inicial("A")
contro.personas_mayor_saldo()
```

```python
class Hora():
  __horas = 0
  __minutos=0
  __segundos = 0
  def __init__(self, horas, minutos, segundos):
    self.setHora(horas, minutos, segundos)
  def setHora(self, horas, minutos, segundos):
    if 0 <= horas <= 23:
      self.__horas = horas
    else:
      self.__horas = 0
    if 0 <= minutos <= 59:
      self.__minutos = minutos
    else:
      self.__minutos = 0
    if 0 <= segundos <= 59:
      self.__segundos = segundos
    else:
      self.__segundos = 0
  def getHora(self):
    return [self.__horas, self.__minutos, self.__segundos]
  def imprimirHora(self):
    print(f"{self.__horas}:{self.__minutos}:{self.__segundos}")
  def getHoras(self):
    return self.__horas
  def getMinutos(self):
    return self.__minutos
  def getSegundos(self):
    return self.__segundos
  def setHoras(self, horas):
    self.__horas = horas
  def setMinutos(self, minutos):
    self.__minutos = minutos
  def setSegundos(self, segundos):
    self.__segundos = segundos
  def incrementarSegundo(self):
    self.__segundos +=1
    if self.__segundos == 60:
      self.__segundos = 0
      self.__minutos += 1
      if self.__minutos == 60:
        self.__minutos = 0
        self.__horas += 1
        if self.__horas == 24:
          self.__horas = 0
  def getSegundosTotales(self):
    return self.__segundos + self.__minutos * 60  + self.__horas * 60 * 60

class Controladora():
  __lista  = []
  def agregar(self, hora):
    self.__lista.append(hora)
  def mostrarHoras(self):
    for hora in self.__lista:
      hora.imprimirHora()
  def incrementarUnSegundo(self):
    for hora in self.__lista:
      hora.incrementarSegundo()
  def segundosTotalesAHoras(self, segundos):
    horas = segundos // 3600
    segundos = segundos % 3600
    minutos = segundos // 60
    segundos = segundos % 60
    return horas, minutos, segundos
  def restoDelDia(self):
    for hora in self.__lista:
      segundosRestantes = 24*60*60 - hora.getSegundosTotales()
      horas, minutos, segundos = self.segundosTotalesAHoras(segundosRestantes)
      print(f"{horas}:{minutos}:{segundos}")
  def mayorMenorHora(self):
    listaSegundos = []
    for hora in self.__lista:
      listaSegundos.append(hora.getSegundosTotales())
    minSegundos = min(listaSegundos)
    maxSegundos = max(listaSegundos)
    print("Maxima hora:", self.segundosTotalesAHoras(maxSegundos))
    print("Minima hora:", self.segundosTotalesAHoras(minSegundos))

#main

hora = 0
contro = Controladora()
while hora != -1:
  hora = int(input("Ingrese hora: "))
  if hora != -1:
    minuto = int(input("Ingrese minuto: "))
    segundo = int(input("Ingrese segundo: "))
    objHora = Hora(hora, minuto, segundo)
    contro.agregar(objHora)
contro.mostrarHoras()
contro.incrementarUnSegundo()
contro.mostrarHoras()
contro.restoDelDia()
contro.mayorMenorHora()
```
