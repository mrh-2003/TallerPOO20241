```python
def factorial(num):
  fac = 1
  for i in range(1, num + 1):
    fac *= i
  return fac
def calcularSumatoria(n):
  suma = 0
  for i in range(1, n + 1):
    suma += i ** 2 / (2 * factorial(i))
  print(suma)
n = -1
while not(n > 0 and n < 120):
  n = int(input("Ingrese n: "))
calcularSumatoria(n)
```

```python
def graficar(n):
  for fil in range(1, n + 1):
    for esp in range(1, n - fil + 1):
      print(end= "  ")
    for col in range(1, fil * 2):
      if col == 1 or col == fil * 2 - 1:
        if fil % 2 != 0:
          print(end="- ")
        else:
          print(end="| ")
      else:
        print(end="* ")
    print()
n = -1
while not(n >= 4 and n <= 15):
  n = int(input("Ingrese n: "))
graficar(n)
```

```python
def calcularSumatoria(n, b):
  suma = 0
  cont = 4
  for i in range(1, n + 1):
    suma += (-1) ** (i + 1) * i * 2 *b / cont
    cont += 3
  return suma

n = -1
while not(n > 0):
  n = int(input("Ingrese n: "))
b = -1
while not(b >= 1 and b <= 5): # 1<= b <= 5
  b = float(input("Ingrese b: "))
print("La suma es: ", calcularSumatoria(n, b))
```

```python
import random as r
def generar(n):
  lista = []
  for i in range(n):
    lista.append(r.randint(1, 9))
  return lista
def mostrar(lista):
  for i in lista:
    print(i, end=" ")
  print()
def contarRepeticiones(lista):
  for i in range(1, 10):
    print(f"Cantidad de repeticiones para {i}:", lista.count(i))
def esPrimo(num):
  cont = 0
  for i in range(1, num):
    if num % i == 0:
      cont += 1
  return cont == 1
def remplazarValor(lista):
  for i in range(len(lista)):
    if esPrimo(lista[i]):
      lista[i]+=1
  mostrar(lista)
n = -1
while not(n > 0 and n <= 40):
  n = int(input("Ingrese n: "))
lista = generar(n)
mostrar(lista)
contarRepeticiones(lista)
remplazarValor(lista)
```
