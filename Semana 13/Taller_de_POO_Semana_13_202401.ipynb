```python
import pandas as pd

def calcularBalance(df, meses):
  df["Balance"] = df["Ventas"] - df["Gastos"]
  return df[df["Meses"].isin(meses)]["Balance"].sum()

meses = ["Enero", "Febrero", "Marzo", "Abril"]
ventas = [30500, 35600, 28300, 33900]
gastos = [22000, 23400, 18100, 20700]

df = pd.DataFrame({
    "Meses": meses,
    "Ventas": ventas,
    "Gastos": gastos
})
print(calcularBalance(df,["Enero","Abril", "Diciembre"]))

```

```python
import pandas as pd
def extraerInformacion():
  df = pd.read_excel("cotizacion.xlsx")
  df = df[["Final",	"Mximo",	"Mnimo",	"Volumen",	"Efectivo"]]
  return pd.DataFrame([df.max(), df.min(), df.mean(), df.median()], index=["Maximo", "Minimo", "Media", "Mediana"])

print(extraerInformacion())
```
