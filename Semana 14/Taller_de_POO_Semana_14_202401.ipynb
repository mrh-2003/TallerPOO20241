```python
import pandas as pd
import matplotlib.pyplot as plt

def graficar(serie, titulo):
  serie.plot(kind='pie', title= titulo)
  plt.savefig(titulo+'.png')
  plt.show()

ventas = {'Enero': 6000, 'Febrero': 6542, 'Marzo': 9852}
serie = pd.Series(ventas)
graficar(serie, "Ventas del primer trimestre")
```

```python
import pandas as pd
import matplotlib.pyplot as plt

def graficar(df):
  df.plot(kind='line', title= 'Evolución de ingresos y gastos', ylim=(0, 70))
  plt.show()

ventas = {
    'Ingresos': [15, 60, 20, 30],
    'Gastos': [25, 30, 20, 10],
    'Meses': ['Enero', 'Febrero', 'Marzo', 'Abril']
}
df = pd.DataFrame(ventas).set_index('Meses')
graficar(df)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv('tita.csv')

df["Survived"].value_counts().plot(kind= 'pie', ylabel='', autopct='%.2f %%', labels=['Muertos', 'Supervivientes'], title= "Fallecidos y sobrevivientes")
plt.show()
```

```python
df["Age"].plot(kind= 'hist',  title= "Histograma de edades")
plt.show()
```

```python
df["Pclass"].value_counts().plot(kind= 'bar', title= "Cantidad de personas por clase", xlabel='Clase', ylabel= 'Cantidad')
plt.show()
```

```python
df.groupby(['Survived', 'Pclass']).size().unstack().plot(kind='bar', title='Sobrevivientes por clases')
plt.show()
```

```python
df.groupby(['Survived', 'Pclass']).size().unstack().plot(kind='bar',stacked=True, title='Sobrevivientes por clases')
plt.show()
```
