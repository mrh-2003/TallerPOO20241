```python
import math as m

def calcularAreaVolumen(arista):
  area = m.pow(arista, 2) * m.sqrt(3)
  volumen = m.sqrt(2)* arista ** 3 / 12
  return area, volumen

#main
arista = float(input("Ingrese la arista: "))
area, volum = calcularAreaVolumen(arista)
print("El area es:", round(area, 2))
print("El volumen es:", volum)
```

```python
def mostrarDatos(codigo): # 15621212,04
  ss = codigo % 100 #04
  codigo =  codigo // 100 # 156212,12
  mm = codigo % 100
  codigo =  codigo // 100 # 1562,12
  hh = codigo % 100
  tttt =  codigo // 100 # 1562
  print(f"Codigo del trabajador: {tttt}")
  print("Hora de entrada:", hh)
  print("Minuto de entrada:", mm)
  print("Segundo de entrada:", ss)
codigo = -1
while not(len(str(codigo)) == 10):
  codigo = int(input("Ingrese codigo: "))
mostrarDatos(codigo)
```
