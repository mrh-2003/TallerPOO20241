```python
import random as r
class Persona():
  __nombre = ""
  __edad = 0
  __dni = ""
  __sexo = "H"
  __peso = 0.0
  __altura = 0.0
  def __init__(self, nombre, edad, sexo, peso, altura):
    self.__nombre = nombre
    self.__edad = edad
    self.__sexo = sexo
    self.__peso = peso
    self.__altura = altura
    self.generarDNI()

  def getNombre(self):
    return self.__nombre
  def setNombre(self, nombre):
    self.__nombre = nombre
  def getEdad(self):
    return self.__edad
  def setEdad(self, edad):
    self.__nombre = edad
  def getDNI(self):
    return self.__dni
  def setdni(self, dni):
    self.__dni = dni
  def getsexo(self):
    return self.__sexo
  def setsexo(self, sexo):
    self.__sexo = sexo
  def getpeso(self):
    return self.__peso
  def setpeso(self, peso):
    self.__peso = peso
  def getaltura(self):
    return self.__altura
  def setaltura(self, altura):
    self.__altura = altura
  def calcularIMC(self):
    imc = self.__peso / (self.__altura ** 2)
    if imc < 20:
      return -1
    elif imc <= 25:
      return 0
    else:
      return 1
  def esMayorEdad(self):
    return self.__edad >= 18
  def generarDNI(self):
    self.__dni = str(r.randint(10000000, 99999999))
  def getInfo(self):
    return f"Nombre: {self.__nombre}\nDNI:{self.__dni}\nEdad: {self.__edad}\nSexo: {self.__sexo}\nPeso: {self.__peso}\nAltura: {self.__altura}"


#main

def main():
  nombres = ["Luis", "Xime", "Jose", "Percy"]
  apellidos = ["Quispe", "Mamani", "Flores", "Mendoza"]
  #nombre = ""
  #while not(len(nombre) > 0):
  #  nombre = input("Ingrese nombre: ")
  nombre = nombres[r.randint(0, 3)] + " " + apellidos[r.randint(0, 3)]
  edad = -1
  while not(edad>=0):
    edad = int(input("Ingrese edad: "))
  #dni = input("Ingrese dni: ")
  peso = float(input("Ingres el peso: "))
  altura = float(input("Ingres el altura: "))
  sexo = input("Ingrese sexo: ")
  persona = Persona(nombre, edad, sexo, peso, altura)
  imc = persona.calcularIMC()
  if imc == -1:
    print(nombre,"debajo de su peso ideal")
  elif imc == 0:
    print(nombre,"peso ideal")
  else:
    print( nombre, "tienes sobrepeso")
  print("Es mayor de edad?: ", persona.esMayorEdad())
  print(persona.getInfo())
main()
```

```python
import random as r
class Contrasenia():
  __longitud = 8
  __contra = ""
  def __init__(self, longitud):
    self.__longitud = longitud
    self.generarContra()
  def esSeguro(self):
    cNum = cMayus = cMinus =0
    for caracter in self.__contra:
      if caracter.isdigit():
        cNum+= 1
      elif caracter == caracter.upper():
        cMayus+=1
      else:
        cMinus+=1
    return cNum > 5 and cMayus > 2 and cMinus > 1
  def generarContra(self):
    for i in range(self.__longitud):
      op = r.randint(0, 2)
      if op == 0:
        self.__contra += chr(r.randint(65, 90))
      elif op == 1:
        self.__contra += chr(r.randint(97, 122))
      else:
        self.__contra += str(r.randint(0, 9))
  def getContra(self):
    return self.__contra
  def getLongitud(self):
    return self.__longitud
  def setContra(self, contra):
    self.__contra = contra

class Controladora():
  __contrasenias = []
  __sonFuertes = []
  def generar(self):
    n = r.randint(6, 12)
    for i in range(n):
      self.__contrasenias.append(Contrasenia(r.randint(15, 20)))
  def analizar(self):
    for contra in self.__contrasenias:
      self.__sonFuertes.append(contra.esSeguro())
  def imprimir(self):
    for i in range(len(self.__contrasenias)):
      print(f"Contraseña {i + 1} es {self.__contrasenias[i].getContra() } es fuerte: { 'V' if self.__sonFuertes[i] else 'F' }")

def main():
  control = Controladora()
  control.generar()
  control.analizar()
  control.imprimir()
main()
```
